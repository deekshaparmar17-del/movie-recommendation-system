```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
data = {
    "title": ["Avatar", "Titanic", "Inception", "The Dark Knight", "Interstellar", "Avengers", "Joker"],
    "genres": [
        "Action Adventure", "Drama Romance", "Sci-Fi Thriller",
        "Action Crime", "Sci-Fi Adventure", "Action Superhero", "Crime Drama"
    ],
    "overview": [
        "A marine on an alien planet joins the indigenous people.",
        "A tragic love story set on the sinking Titanic.",
        "A skilled thief enters people's dreams to steal secrets.",
        "A vigilante battles crime in Gotham City.",
        "Explorers travel through a wormhole in space.",
        "Earth's mightiest heroes team up to fight a powerful enemy.",
        "A mentally troubled man becomes a criminal mastermind."
    ]
}

df = pd.DataFrame(data)
df
```

```python
df["combined"] = df["genres"] + " " + df["overview"]
df["combined"] = df["combined"].fillna("")
df.head()
```

```python
vectorizer = TfidfVectorizer(stop_words='english')
vectors = vectorizer.fit_transform(df["combined"])
vectors.shape
```

```python
def recommend(movie_title, n=3):
    if movie_title not in df["title"].values:
        return ["Movie not found in the dataset."]
    
    index = df[df["title"] == movie_title].index[0]
    similarity_scores = cosine_similarity(vectors[index], vectors).flatten()
    similar_indices = similarity_scores.argsort()[::-1][1:n+1]
    
    return df["title"].iloc[similar_indices].tolist()
```

```python
recommend("Avatar")
```

```python
recommend("Inception")
```

```python
recommend("Joker")
```
